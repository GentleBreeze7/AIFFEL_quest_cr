```python
class Square:
    def __init__(self):
        while True:
            try:
                self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n1. 직사각형\n2. 평행사변형\n3. 사다리꼴\n>>> '))
                if self.square in [1, 2, 3]:
                    break
                else:
                    print("1, 2, 3 중에서 선택해주세요.")
            except ValueError:
                print("숫자를 입력해주세요.")

        if self.square == 1:
            print('직사각형의 넓이를 구합니다.')
            print(self.rect())
        elif self.square == 2:
            print('평행사변형의 넓이를 구합니다.')
            print(self.par())
        elif self.square == 3:
            print('사다리꼴의 넓이를 구합니다.')
            print(self.trape())

    def rect(self):
        while True:
            try:
                width, vertical = map(float, input('가로와 세로를 입력하세요 (예: 가로,세로): ').split(','))
                area = width * vertical
                return f'직사각형의 넓이는: {area}'
            except ValueError:
                print("올바른 형식으로 입력해주세요 (예: 5,4).")

    def par(self):
        while True:
            try:
                base, height = map(float, input('밑변과 높이를 입력하세요 (예: 밑변,높이): ').split(','))
                area = base * height
                return f'평행사변형의 넓이는: {area}'
            except ValueError:
                print("올바른 형식으로 입력해주세요 (예: 5,4).")

    def trape(self):
        while True:
            try:
                base1, base2, height = map(float, input('밑변1, 밑변2, 높이를 입력하세요 (예: 밑변1,밑변2,높이): ').split(','))
                area = (base1 + base2) * height / 2
                return f'사다리꼴의 넓이는: {area}'
            except ValueError:
                print("올바른 형식으로 입력해주세요 (예: 5,6,4).")


# 클래스 사용
a = Square()
```

```python
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_menu = []  # 주문한 음료 리스트
        self.order_price = []  # 주문한 가격 리스트
        self.price_sum = 0  # 합계 금액

    def menu_print(self):
        print("\n--- 메뉴판 ---")
        for i in range(len(self.menu)):
            print(f"{i + 1} {self.menu[i]}  :  {self.price[i]} 원")

    def menu_select(self):
        while True:
            print()
            try:
                n = int(input("음료 번호를 입력하세요 (지불: 0): "))

                if n == 0:  # 지불 선택
                    print("\n주문이 완료되었습니다.")
                    print("주문 리스트:", self.order_menu)
                    print("가격 리스트:", self.order_price)
                    print(f"총 결제 금액: {self.price_sum}원")
                    break

                if 1 <= n <= len(self.menu):
                    # 음료 온도 선택
                    t = 0
                    while t not in [1, 2]:
                        t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요: "))
                        if t == 1:
                            temp = "HOT"
                        elif t == 2:
                            temp = "ICE"
                        else:
                            print("1과 2 중 하나를 입력하세요.")

                    # 주문 추가
                    self.order_menu.append(f"{temp} {self.menu[n - 1]}")
                    self.order_price.append(self.price[n - 1])
                    self.price_sum += self.price[n - 1]

                    print(f"추가 주문 음료 {temp} {self.menu[n - 1]} : {self.price[n - 1]} 원")
                    print(f" 합계 :  {self.order_price} 원")

                else:
                    print("없는 메뉴입니다. 다시 주문해 주세요.")

            except ValueError:
                print("유효한 숫자를 입력해 주세요.")

# 키오스크 실행
def run_kiosk():
    kiosk = Kiosk()
    kiosk.menu_print()
    kiosk.menu_select()

run_kiosk()
```

```python
class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]
        self.order_menu = []  # 주문한 음료 리스트
        self.order_price = []  # 주문한 가격 리스트
        self.price_sum = 0  # 합계 금액

    def menu_print(self):
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')
        for i in range(len(self.menu)):
            print(f"{i + 1} {self.menu[i]}  :  {self.price[i]} 원")
        for i in range(5):
            print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')

    def menu_select(self):
        while True:
            print()
            try:
                n = int(input("음료 번호를 입력하세요 (지불: 0): "))

                if n == 0:  # 지불 선택
                    self.table()
                    self.pay()
                    break

                if 1 <= n <= len(self.menu):
                    # 음료 온도 선택
                    t = 0
                    while t not in [1, 2]:
                        t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요: "))
                        if t == 1:
                            temp = "HOT"
                        elif t == 2:
                            temp = "ICE"
                        else:
                            print("1과 2 중 하나를 입력하세요.")

                    # 주문 추가
                    self.order_menu.append(f"{temp} {self.menu[n - 1]}")
                    self.order_price.append(self.price[n - 1])
                    self.price_sum += self.price[n - 1]

                    print(f"추가 주문 음료 {temp} {self.menu[n - 1]} : {self.price[n - 1]} 원")
                    print(f" 합계 :  {self.price_sum} 원")

                else:
                    print("없는 메뉴입니다. 다시 주문해 주세요.")

            except ValueError:
                print("유효한 숫자를 입력해 주세요.")

    def pay(self):
        while True:
            method = input("지불 방법을 선택하세요 (cash 또는 card): ").lower()
            if method in ['cash', '1']:
                print("직원을 호출하겠습니다.")
                break
            elif method in ['card', '2']:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                break
            else:
                print("잘못된 입력입니다. 다시 시도해주세요.")

    def table(self):
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문한 상품명과 가격 출력
        for i in range(len(self.order_menu)):
            print(f"{self.order_menu[i]} : {self.order_price[i]}원")

        print(f"합계 금액 : {self.price_sum}원")

        for i in range(5):
            print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')

# 키오스크 실행
def run_kiosk():
    kiosk = Kiosk()
    kiosk.menu_print()
    kiosk.menu_select()

run_kiosk()
```
